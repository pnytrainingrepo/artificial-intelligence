```python
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
```

```python
# Create a simple dataset
X, y = make_moons(n_samples=5000, noise=0.2, random_state=42)
```

```python
print(X)
```

```python
print(y)
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create a Neural Network model using MLPClassifier
nn_model = MLPClassifier(hidden_layer_sizes=(128, 64, 32), max_iter=5000, random_state=42)
```

```python
# Train the model
nn_model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = nn_model.predict(X_test)
```

```python
# Evaluate the model accuracy
accuracy = accuracy_score(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
print("Classification Report:\n", class_report)
```

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
# Load the digits dataset
digits = load_digits()

# Prepare the features (image data) and labels
X = digits.data  # 8x8 images flattened into 64 features
y = digits.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a Neural Network model using MLPClassifier
nn_model = MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=1000, random_state=42)
```

```python
digits.shape()
```

```python
# Train the model
nn_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = nn_model.predict(X_test)

# Evaluate the model accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
